```python
from pathlib import Path
root_path = Path("../logs_recognition/size")
```

```python
csv_paths = ['csv/mse-13.204.csv', 'csv/mse-15.637.csv', 'csv/mse-42.169.csv', 'csv/mse-43.829.csv', 'csv/mse-9.855.csv']
angular_size = ['1.9e-4', '1.2e-4', '9e-5', '7e-5', '1e-3']
mae = ['0.501', '0.613', '1.291', '1.36', '0.346']
```

```python
for index, csv_path in enumerate(csv_paths):
    df = pd.read_csv(csv_path, index_col=0)
    fig, ax = plt.subplots(figsize=(7, 7))
    palette = sns.color_palette('mako_r', n_colors=13)
    sns.violinplot(x=df.Real.astype(int), y=df.Pred, inner=None, ax=ax, palette=palette)
    newax = fig.add_axes(ax.get_position(), frameon=False)
    x = np.arange(64, 75)
    y = x
    # with sns.set_theme(style="darkgrid"):
    newax.plot(x, y, '-', color='#dcbe87', markersize=6, lw=2)
    newax.plot(x, y, 'o', color='#FFC75F', markersize=6, lw=2)
    newax.grid(False)
    # ax.set_xlim(60, 77)
    ax.set_ylim(60, 77)
    newax.set_ylim(60, 77)
    newax.set_xticklabels([])
    newax.set_yticklabels([])
    if index !=2:
        ax.set_yticklabels([])
    else:
        ax.set_ylabel('Predicted distribution of the size (px)')
    ax.set_xlabel('Real size of black hole (px)')
    
    # plt.tight_layout()
    plt.savefig(f'images/violin_{angular_size[index]}_{mae[index]}.png', dpi=600)
    plt.savefig(f'images/violin_{angular_size[index]}_{mae[index]}.pdf')
```
