```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5.5), sharey=True)
for index, csv_path in enumerate(csv_paths):
    ax = axes[index]
    # ax.axis('square')
    df = pd.read_csv(csv_path, index_col=0)
    palette = sns.color_palette('mako_r', n_colors=13)
    sns.violinplot(x=df.Real.astype(int), y=df.Pred, inner=None, ax=ax, palette=palette)
    # fig.tight_layout()
    # ax.set_xlim(60, 77)
    ax.set_ylim(60, 77)
    ax.set_box_aspect(1)
    if index !=0:
        # ax.set_yticklabels([])
        ax.set_ylabel('')
    else:
        ax.set_ylabel('Predicted distribution of the size (px)')
    ax.set_xlabel(rf'Real size, MAE$={mae[index]:.1f}, \theta_{{AD}}={angular_sizes[index]*64/3.87e-3:.2f}\theta_c$')
    # ax.set_xticks(np.arange(64, 74, 2))

fig.tight_layout()
for index, csv_path in enumerate(csv_paths):
    ax = axes[index]
    df = pd.read_csv(csv_path, index_col=0)
    newax = fig.add_axes(ax.get_position(), frameon=False)
    x = np.arange(64, 75)
    y = x
    # with sns.set_theme(style="darkgrid"):
    newax.plot(x, y, '-', color='#FFC75F', markersize=6, lw=2, alpha=0.8)
    # newax.plot(x, y, '-', color='#dcbe87', markersize=6, lw=2, alpha=0.5)
    newax.plot(x, y, 'o', color='#FFC75F', markersize=6, lw=2)
    newax.set_ylim(60, 77)
    newax.grid(False)
    newax.set_xticklabels([])
    newax.set_yticklabels([])
    
plt.savefig(f'images/violin4.png', dpi=600)
plt.savefig(f'images/violin4.pdf')
```
