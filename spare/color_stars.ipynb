```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2
import cmocean
import os
from PIL import Image
```

# Before

```python
img_arr = cv2.imread('../stars/BHs.png', 0)
img_arr = cv2.resize(img_arr, (1024, 1024))
img_arr = img_arr.astype('float64')
img_colored = np.zeros((img_arr.shape[0], img_arr.shape[1], 3), dtype=np.float64)
b, g, r = 0.93, 1, 1
img_colored[:, :, 0] = img_arr * b
img_colored[:, :, 1] = img_arr * g
img_colored[:, :, 2] = img_arr * r
# plt.imshow(img_colored)
img_colored = cv2.resize(img_colored, (1024, 1024))
cv2.imwrite('conv_show/colored_before.png', img_colored)
# cv2.imshow('colored', img_colored)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
# plt.show()
```

```python
with open('../stars/labels.txt') as f:
    xc, yc, w, h = map(float, f.readline().strip().split(' ')[1:])
    for i in range(4):
        f.readline()
    xc2, yc2, w2, h2 = map(float, f.readline().strip().split(' ')[1:])
xc, yc, w, h = int(xc * 1024), int(yc * 1024), int(w * 1024), int(h * 1024)
xc2, yc2, w2, h2 = int(xc2 * 1024), int(yc2 * 1024), int(w2 * 1024), int(h2 * 1024)

fig, ax = plt.subplots(figsize=(7, 7))
cmap = cmocean.cm.matter
# plt.imshow(img_arr)
ax.imshow(img_arr, cmap='gray')
x1, x2, y1, y2 = xc - w // 2, xc + w // 2, yc - h // 2, yc + h // 2
x12, x22, y12, y22 = xc2 - w2 // 2, xc2 + w2 // 2, yc2 - h2 // 2, yc2 + h2 // 2
axin1 = ax.inset_axes(
    [0.85, 0.85, 0.1, 0.1],
    xlim=(x1, x2), ylim=(y1, y2), xticklabels=[], yticklabels=[], xticks=[], yticks=[])

axin2 = ax.inset_axes(
    [0.05, 0.35, 0.1, 0.1],
    xlim=(x12, x22), ylim=(y12, y22), xticklabels=[], yticklabels=[], xticks=[], yticks=[])

axin1.imshow(img_arr, cmap='gray')
axin1.spines['bottom'].set_color('white')
axin1.spines['top'].set_color('white') 
axin1.spines['right'].set_color('white')
axin1.spines['left'].set_color('white')
ax.indicate_inset_zoom(axin1, edgecolor="white")

axin2.imshow(img_arr, cmap='gray')
axin2.spines['bottom'].set_color('white')
axin2.spines['top'].set_color('white') 
axin2.spines['right'].set_color('white')
axin2.spines['left'].set_color('white')
ax.indicate_inset_zoom(axin2, edgecolor="white")


# Hide grid lines
ax.grid(False)
# Hide axes ticks
ax.set_xticks([])
ax.set_yticks([])
fig.tight_layout()
plt.savefig('conv_show/zoomed_before.png', dpi=600)
plt.savefig('conv_show/zoomed_before.pdf', dpi=600)
```
