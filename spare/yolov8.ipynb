```python
import os
import glob as glob
import matplotlib.pyplot as plt
import numpy as np
import datetime
from ultralytics import YOLO
import comet_ml
```

```python
TRAIN = True
EPOCHS = 300
star = 100
BH = 10
num_photo = 500
batch_size = 16
size = 800
data_dir = f"stars{star}_BH{BH}_num{num_photo}"
```

```python
os.environ["COMET_MAX_IMAGE_PREDICTIONS"] = "100"
now = datetime.datetime.now()
date_string = now.strftime("%Y-%m-%d_%H-%M-%S")
project_name = f"yolov8-{date_string}"
curr_dir = f"logs_yolo/{date_string}"
comet_ml.init(project_name=project_name)
np.random.seed(2024)
```

```python
model = YOLO('logs_yolo/yolov8-2024-01-18_13-00-43/train/weights/epoch270.pt')  # load a pretrained model (recommended for training)

results = model.train(
    data=f'datasets/{data_dir}/datav8.yaml',
    project=f"logs_yolo/{project_name}",
    batch=batch_size,
    save_period=30,
    save_json=True,
    epochs=EPOCHS,
    imgsz=size
)
```

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```
