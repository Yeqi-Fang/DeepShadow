```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../tele_datasets/224/20240104143626_07df859daba249f5b87a33a959a410fe.png')
plt.imshow(img)
plt.show()
```

```python
BH_size = 224
```

```python
centerX, centerY = (BH_size // 2, BH_size // 2)
angle = 180.32
M = cv2.getRotationMatrix2D((centerX, centerY), angle, 1.0)
rotate = cv2.warpAffine(img, M, (BH_size, BH_size))
plt.imshow(rotate)
plt.show()
```
